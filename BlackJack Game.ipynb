```python
from random import shuffle

suits = ('Hearts', 'Diamonds', 'Spades', 'Clubs')
ranks = ('Two', 'Three', 'Four', 'Five', 'Six', 'Seven', 'Eight', 'Nine', 'Ten', 'Jack', 'Queen', 'King', 'Ace')
values = {'Two':2, 'Three':3, 'Four':4, 'Five':5, 'Six':6, 'Seven':7, 'Eight':8, 'Nine':9, 'Ten':10, 'Jack':10,
         'Queen':10, 'King':10, 'Ace':11}
playing = True
```

```python
class CARD:
    def __init__(self, suit, rank):
        self.suit = suit
        self.rank = rank

    def __str__(self):
        return self.rank + " of " + self.suit
```

```python
class DECK:
    def __init__(self):
        self.all_52_cards = []
        for suit in suits:
            for rank in ranks:
                self.all_52_cards.append(CARD(suit, rank))

    def __str__(self):
        avail_cards = ""
        for card in self.all_52_cards:
            avail_cards += "\n" + str(card)
        return avail_cards
    
    def shuffle(self):
        shuffle(self.all_52_cards)

    def remove_card(self):
        return self.all_52_cards.pop()
        
```

```python
class HAND:
    def __init__(self):
        self.cards = []
        self.value = 0
        self.aces = 0

    def add_card(self, card):
        self.cards.append(card)
        self.value += values[card.rank]
        if card.rank == 'Ace':
            self.aces += 1

    def adjust_aces(self):
        for card in self.cards:
            if card.rank == 'Ace' and self.value > 21:
                card.value -= 10
                self.aces -= 1
            
```

```python
class CHIPS:
    def __init__(self):
        self.total = 100
        self.bet = 0

    def win_bet(self):
        self.total += self.bet

    def lose_bet(self):
        self.total -= self.bet
```

```python
def take_player_bet(chips):
    
    while True:
        try:
            chips.bet = int(input('Enter your bet amount:'))
        except:
            print("Please enter valid input!!")
        else:
            if chips.bet < chips.total:
                break
            else:
                print("You have exceeded your balance, you have {} chips. Try again!!".format(chips.total))
```

```python
def hit(deck, hand):
    hand.add_card(deck.remove_card())
    hand.adjust_aces()
    
```

```python
def hit_or_stand(deck, hand):
    global playing

    while True:
        
        hit_stand = input("Do you want to hit or stand (enter 'h' or s): ")
   
        if hit_stand[0].lower() == 'h':
            hit(deck, hand)
            break
        elif hit_stand[0].lower() == 's':
            playing = False
            break
        else:
            print("Incorrect value provided!!, enter 'h' for hit and 's' for stand.")
            continue
                
```
